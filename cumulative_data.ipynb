Calculate cumulative number of cases

```python
# import dependencies
import pandas as pd
```

```python
# set province
province = "CATANDUANES"
```

```python
# load csv file# read from csv file
df = pd.read_csv(f"aggregated_data/{province}.csv")
df.head()
```

```python
# add new column for change in cases
df["d_cases"] = df["NewCases"] - df["Deaths"] - df["Recoveries"]

df
```

```python
# calculate cumulative sum
df["n"] = df["d_cases"].cumsum()
df
```

```python
# export to csv

df.to_csv(f"aggregated_data/{province}.csv")
```
