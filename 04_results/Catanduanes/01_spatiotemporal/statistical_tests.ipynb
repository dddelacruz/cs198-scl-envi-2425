```python
import pandas as pd

df = pd.read_csv('./Catanduanes_results.csv')
df.head()
```

```python
from scipy.stats import kruskal

grouped = [group['typeII'].values for _, group in df.groupby('Municipality')]

h, p = kruskal(*grouped)

print(f"Kruskal-Wallis H-statistic: {h:.3f}, p-value: {p:.4f}")
```

```python
import scikit_posthocs as sp

dunn = sp.posthoc_dunn(df, val_col='typeII', group_col='Municipality', p_adjust='bonferroni')

dunn.to_csv('./Catanduanes_Dunn_pairwise_comparison.csv')
```
