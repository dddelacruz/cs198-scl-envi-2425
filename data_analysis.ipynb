```python
dataset['CityMunRes'].value_counts()
```

```python
import seaborn as sns
from matplotlib import rcParams
from matplotlib import pyplot as plt

ds = dataset['CityMunRes'].value_counts().to_dict()
xs = [key for key, _ in ds.items()]
ys = [value for _, value in ds.items()] 

rcParams['figure.figsize'] = 15, 9
plt.xticks(rotation=45)
sns.barplot(x=xs, y=ys)
```

```python
sex = dataset['Sex'].value_counts().to_dict()
xs = [key for key, _ in sex.items()]
ys = [value for _, value in sex.items()] 

rcParams['figure.figsize'] = 5, 5
plt.pie(ys, labels=xs, autopct="%1.2f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
rem = dataset['RemovalType'].value_counts().to_dict()
xs = [key for key, _ in rem.items()]
ys = [value for _, value in rem.items()] 

rcParams['figure.figsize'] = 5, 5
plt.pie(ys, labels=xs, autopct="%1.2f%%")
plt.title("Removal Type Distribution")
plt.show()
```

```python
import ipywidgets as widgets
from ipywidgets import interact

municipalities = list(agg['Municipality'].unique())

def plot_time_series(municipality):
    print(municipality)
    filtered = agg[agg['Municipality'] == municipality].copy()
    plt.plot(filtered['Date'], filtered['NewCases'])

# print(municipalities)
rcParams['figure.figsize'] = 16,9
x = widgets.Dropdown(options=municipalities)
interact(plot_time_series, municipality=x)
plt.show()
```

```python
dataset.head()
```

```python
time_series_data = dataset[['DateRepConf', 'CityMunRes']].groupby(['DateRepConf','CityMunRes']).value_counts().unstack().copy()
```

```python
time_series_data.head()
```

```python
time_series_data.fillna(0, inplace=True)
```

```python
time_series_data.columns
```

```python
time_series_data['DateRepConf'] = time_series_data.index
```

```python
def updated_plot_time_series(municipality):
    # print(municipality)
    # filtered = agg[agg['Municipality'] == municipality]
    plt.plot(time_series_data['DateRepConf'], time_series_data[municipality])

municipalities = list(time_series_data.columns)[:-1]

rcParams['figure.figsize'] = 16,9
x = widgets.Dropdown(options=municipalities)
interact(updated_plot_time_series, municipality=x)
plt.show()
```
